```python
# encode the target variable for visualization 
# the class represents whether the applicant has good or bad credit
# these are the actuals where a portion will be used to test the model by checking against these values during training

# encode to turn good/bad into 1 or 0
from sklearn.preprocessing import LabelEncoder
pd.set_option('display.max_columns',25)
df["class"] = LabelEncoder().fit_transform(df["class"])
```

```python
'''
You MUST include at least four (4) numeric features and at least three (3) nominal features.
numeric (4)
nominal = categorical (3) but there is no intrinsic ordering to the categories(can be numbers but ordering the numbers would be meaning less)
'''

numeric = df[['age', 'duration', 'credit_amount', 'existing_credits','num_dependents']]
categorical = df[['credit_history', 'purpose', 'housing', 'job']]
```

```python
categorical
```

```python
'''
Step 4

Preprocessing
Perform any needed preprocessing on the chosen features, including:

    Scaling
    Encoding
    Dealing with NaN values

Note:
Use only the preprocessing steps you think are useful. 
'''
# encode the categorical variables
categorical = pd.get_dummies(categorical)
categorical
```

```python
credit = pd.concat([numeric,categorical],axis = 1)
credit
```

```python
'''
Step 5

Splitting the Data
Split your data as follows:
80% training set
10% validation set
10% test set 
'''
y = df['class'].values

x_columns = len(credit.columns)

x = credit.iloc[:, 0:x_columns].values
```

```python
from sklearn.model_selection import train_test_split
# test_size = 0.2 means reserve 20% of the dataset for testing, leaving 80% for the training
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# test_size = 0.5 is 50% or the original test size (0.2) to get 10% of the total set
X_validate, X_test, y_validate, y_test = train_test_split(X_test, y_test, test_size = 0.5, random_state = 0)
```

```python
print(len(X_train), len(X_test), len(X_validate))
```

```python
'''
Step 6

Training Classifiers
Use the KNN-classifier model to train your data.
Choose the best k for the k-nearest neighbor (KNN) algorithm by trying different values and validating performance on the validation set.
Note: choosing the best k is an example of hyper-parameter tuning.
Classification Metrics
Print the accuracy score of your final classifier.
Print the confusion matrix. 
'''
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3).fit(X_train,y_train)
knn.score(X_test,y_test)

```

```python
scores = [] # makes sure to start with an empty list every time
neighbors = range(1,20)
# choose the best k value for KNN algorithm by trying different values and validating performance
# choosing the best k is an example of hyper-parameter tuning
for i in neighbors:
    knn = KNeighborsClassifier(n_neighbors = i).fit(X_train,y_train)
    results = knn.score(X_test,y_test)
    scores.append(results)
print(scores)
```

```python
import matplotlib.pyplot as plt
print(neighbors)
print(scores)
plt.plot(neighbors,scores)
```

```python
#run the best model
bestK =  scores.index(max(scores))
print(bestK, max(scores))
bestK = 13
print(bestK, scores[1])
scores_list = pd.DataFrame(
    {'k': neighbors,
     'scores': scores
    })
print(scores_list)
# index value of max score is is best k value (both 8 and 13 are highest, just pick one)
knn = KNeighborsClassifier(n_neighbors = bestK).fit(X_train,y_train)
```

```python
# print the accuracy score of your classifier
result = knn.score(X_validate,y_validate)
print("accuracy = ", result)
```

```python
# print the confusion matrix
from sklearn.metrics import confusion_matrix

predictions = knn.predict(X_validate)

print("confusion matrix: \n", confusion_matrix(y_validate, predictions) )
```
